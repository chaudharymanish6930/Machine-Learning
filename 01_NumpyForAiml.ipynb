NumPy array internally: numpy.ndarray — N-dimensional array. 1D upto nD

0D - Scaler Value - temp

1D - Audio Signal

2D - Grayscale Images

3D - RGB image collection

```python
arr = np.array(10)  #0D
print(arr.dtype)
```

```python
a = np.array([10, 20, 30, 40, 50])

print(a.dtype)    # int64  — every element 8 bytes
print(a.shape)    # (5,)   — 5 elements, 1D
print(a.ndim)     # 1      — dimensions
print(a.size)     # 5      — total elements
print(a.nbytes)   # 40     — 5 * 8 bytes = 40 bytes
```

```python
# ── Zeros and Ones — NN weight initialization ──────────────
zeros = np.zeros((3, 4))           # 3x4 matrix of zeros
ones = np.ones((2, 3, 4))          # 3D array of ones
identity = np.eye(4)               # 4x4 identity matrix
empty = np.empty((100, 100))       # Uninitialized — fastest allocation

print(zeros)
print(ones)
print(identity)
print(empty)

empty = np.empty((5, 5))
print(empty)
```

```python
# ── Range arrays ─────────────────────
seq = np.arange(0, 10, 2)          # [0, 2, 4, 6, 8] — step 2
evenly = np.linspace(0, 1, 5)      # Equal differences
# Generate 5 evenly spaced numbers between 0 and 1 (inclusive)

logspace = np.logspace(0, 3, 4)    # Equal ratios
# Generate 4 numbers whose exponents are evenly spaced

print(seq)
print(evenly)
print(logspace)
```

```python
# ── Random arrays — ML ─────────────────────
np.random.seed(42)                 # used Reproducibility — ALWAYS set
# Sets the starting state of the random generator.

uniform = np.random.rand(3, 4)     # uniform random between 0 and 1
normal = np.random.randn(1000)     # Standard normal distribution
integers = np.random.randint(0, 100, size=(5, 5))  # Random integers
choice = np.random.choice([1,2,3,4,5], size=10, replace=True)

print(uniform)
print(normal)
print(integers)
print(choice)
```

```python
# ── Full ────────────────────────────
full = np.full((3, 3), 7)          # 3x3 matrix — all 7
fill_like = np.full_like(zeros, 5) # Same shape, all 5
# It automatically copies the shape (and usually the data type) from another array
print(full)
print(fill_like)
```

```python
# ── Reshape — same data, new shape ────────────────────────
import numpy as np
flat = np.arange(12)               # [0, 1, 2, ..., 11]
reshaped = flat.reshape(3, 4)      # 3x4 matrix — same memory!
back = reshaped.reshape(-1)        # -1 matlab auto-calculate karo

print(flat)
print(reshaped)
print(back)

reshaped1 = flat.reshape(6, 2) 
print(reshaped1)
```

```python
# ── Copy vs View — CRITICAL ────────────────────────────────
view = arr[:]                      # View — same memory
copy = arr.copy()                  # Copy — new memory
# view modify kiya → original bhi change! copy is safe.

print(view)
print(copy)
```
