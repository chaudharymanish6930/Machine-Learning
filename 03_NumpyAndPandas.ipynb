```python
#interpolate(method = '')
  ##fills missing values by estimating them using interpolation techniques like linear, polynomial, etc.
    ##Uses the trend or pattern of data to estimate missing values., s useful for numerical or time-series data,
    ##Provides smoother, more accurate fills for graphs, analysis, or machine learning.
```

| Method                 | Description                           | Example                      |
| ---------------------- | ------------------------------------- | ---------------------------- |
| `'linear'` *(default)* | Linearly estimates between two values | Good for even time/data      |
| `'time'`               | Interpolates using time-based index   | Use when index is `datetime` |
| `'polynomial'`         | Fits a polynomial curve               | Use `order=` with it         |
| `'spline'`             | Smooth interpolation like a curve     | Needs SciPy                  |
| `'nearest'`            | Takes nearest known value             | Like rounding                |


```python
#data['column name'].interpolate(limit=1) → limit - max num of consecutive values to be filled
```

What does order do?
It specifies the degree of the polynomial used for interpolation.

Higher order = more complex curve.

Lower order = simpler curve (like a straight line if order=1).

| `order` | Curve Type               | Use Case                                 |
| ------- | ------------------------ | ---------------------------------------- |
| 1       | Linear                   | Simple trends                            |
| 2       | Quadratic                | Slight curves                            |
| 3+      | Polynomial (cubic, etc.) | More complex patterns (e.g. time series) |
