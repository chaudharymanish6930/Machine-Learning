| Method                 | Description                           | Example                      |
| ---------------------- | ------------------------------------- | ---------------------------- |
| `'linear'` *(default)* | Linearly estimates between two values | Good for even time/data      |
| `'time'`               | Interpolates using time-based index   | Use when index is `datetime` |
| `'polynomial'`         | Fits a polynomial curve               | Use `order=` with it         |
| `'spline'`             | Smooth interpolation like a curve     | Needs SciPy                  |
| `'nearest'`            | Takes nearest known value             | Like rounding                |


What does order do?
It specifies the degree of the polynomial used for interpolation.

Higher order = more complex curve.

Lower order = simpler curve (like a straight line if order=1).

| `order` | Curve Type               | Use Case                                 |
| ------- | ------------------------ | ---------------------------------------- |
| 1       | Linear                   | Simple trends                            |
| 2       | Quadratic                | Slight curves                            |
| 3+      | Polynomial (cubic, etc.) | More complex patterns (e.g. time series) |

Correlation tells us the strength and direction of a relationship between two numerical variables.

It answers the question:
"If one value increases, does the other also increase or decrease?"

| Value of `r` | Meaning                      | Interpretation                     |
| ------------ | ---------------------------- | ---------------------------------- |
| +1           | Perfect Positive Correlation | Both variables increase together   |
| 0            | No Correlation               | No relationship                    |
| -1           | Perfect Negative Correlation | One increases, the other decreases |

```python
from scipy.stats import zscore
# Step 1: Calculate Z-score for numeric columns
z_scores = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].apply(zscore)
```

```python
print(z_scores)
```

```python
# Step 2: Detect outliers using threshold |Z| > 3
outliers = data[(z_scores.abs() > 3).any(axis=1)]  # any axis=1 → row-wise check
print(outliers)
```

Why Use Z = ±3 as Threshold?

According to the Empirical Rule (68–95–99.7 Rule) for normally distributed data:

Range	% of Data Included

Z between -1 and 1	≈ 68%

Z between -2 and 2	≈ 95%

Z between -3 and 3	≈ 99.7%

So, only 0.3% of values are expected to lie outside ±3, meaning:

Values beyond Z = ±3 are very rare/extreme → considered outliers.

```python
##IQR - INTERQUARTILE RANGE
  #based on middle values
```

| Step | What It Means                                        |
| ---- | ---------------------------------------------------- |
| Q1   | 25th percentile (1st quartile) – niche ke 25% values |
| Q3   | 75th percentile (3rd quartile) – top ke 25% values   |
| IQR  | Q3 - Q1 → middle 50% values ka range                 |


```python
Q1 = data['sepal_width'].quantile(0.25)
Q3 = data['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR
```

```python
outliers = data[(data['sepal_width'] < Lower_limit) | (data['sepal_width'] > Upper_limit)]
```

```python
#operator - OR
print(outliers)
```

```python
##Z-score Method
##how far the value os from mean in terms of std dev
##Z = (value - mean) / standard deviation
```

```python
# Removing the Outliers
import numpy as np
# Upper bound
upper = np.where(data['sepal_width'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data['sepal_width'] <= (Q1-1.5*IQR))

# Removing the Outliers
data.drop(upper[0], inplace=True)
data.drop(lower[0], inplace=True)
```

```python
accuracy_scores = []

for k in range(1, 20):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    accuracy_scores.append(acc)

# Plot accuracy vs k
import matplotlib.pyplot as plt
plt.plot(range(1, 20), accuracy_scores, marker='o')
plt.title("KNN Accuracy for Different k Values")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the two DataFrames manually based on your image
df1 = pd.DataFrame({
    "Job Function": ["IT", "Customer Support", "Engineering"],
    "Percentage": [19.7, 19.0, 12.0]
})

df2 = pd.DataFrame({
    "Job Function": ["Engineering", "IT", "Consulting"],
    "Percentage": [23.0, 19.2, 8.7]
})

# Set up the side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(10,6), sharey=True)

# Common purple palette
purple_palette = sns.color_palette("Purples_r", n_colors=3)

# Plot Philippines
sns.barplot(data=df1, x="Job Function", y="Percentage", ax=axes[0], palette=purple_palette)
axes[0].set_title("Philippines", fontsize=13)
axes[0].tick_params(axis='x')
for p in axes[0].patches:
    axes[0].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=9, xytext=(0, 3),
                     textcoords='offset points')

# Plot India
sns.barplot(data=df2, x="Job Function", y="Percentage", ax=axes[1], palette=purple_palette)
axes[1].set_title("India", fontsize=13)
axes[1].tick_params(axis='x')
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%',
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=9, xytext=(0, 3),
                     textcoords='offset points')

# Add common title and adjust layout
plt.suptitle("Job Function Distribution: Philippines vs India", fontsize=15)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
