```python
import numpy as np
import time

N=10_00_000

python_list = list(range(N))

start=time.time()
total=sum(python_list)
python_time=time.time()-start

print(f'Python List: {python_time:.4f} seconds')
```

```python
#Time Calculation
import numpy as np

numpy_array = np.arange(N)
start = time.time()
total = numpy_array.sum()
numpy_time = time.time() - start

print(f'NumPy array:  {numpy_time:.4f} seconds')
print(f'NumPy is {python_time / numpy_time:.0f}x faster')
```

```python
#Zeros and ones : used for initialising weights and biases
# np.zeros : a matrix of all zeros
# Used to initialise: bias vectors, output buffers
zeros = np.zeros((3, 4))    # 3 rows, 4 columns
print(zeros)
```

```python
# np.full : fill with any value
mask = np.full((3, 3), fill_value=-1e9)  # used in attention masking
print(mask)
```

```python
#Ranges and generating sequences
# np.arange(start, stop, step) : like Python range() but returns ndarray
arr = np.arange(0, 10, 2)
print(arr)
```

```python
## np.linspace(start, stop, num) : evenly spaced, INCLUDES endpoint
# Used constantly for: learning rate schedules, plotting axes, sin/cos curves
x = np.linspace(0, 1, 5)
print(x)
```

```python
#Random arrays : the heart of ML initialisation
# Always set a seed for reproducibility in experiments
rng = np.random.default_rng(42)

weights = rng.normal(
    0, 0.01, (4,3)
)

print(weights)
```

```python
students = 100

attendance = np.random.randint(
    50,
    100,
    students
)

marks = np.random.randint(
    30,
    100,
    students
)

projects = np.random.randint(
    1,
    10,
    students
)

communication = np.random.randint(
    1,
    10,
    students
)

dataset = np.column_stack(
    (
        attendance,
        marks,
        projects,
        communication
    )
)

print(dataset)
```

```python
score = (
    attendance
    + marks
    + projects*10
    + communication*10
)

top_students = score > 180

print(top_students.sum())
```
